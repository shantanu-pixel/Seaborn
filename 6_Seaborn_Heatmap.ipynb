# Heatmaps

### A heat map is a graphical representation of data where the individual values contined in a matrix are represented as colors. 
### (generally drawn between two categorical data, but sumtime drawn discrete numerical data)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
flights=sns.load_dataset('flights')
```

```python
flights.head()
```

```python
x=flights.pivot_table
(index='year', columns='month', 
                      values='passengers', aggfunc='sum')
```

```python
sns.heatmap(x, cbar=False)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5)
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='viridis')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='summer')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='winter')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='autumn')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='spring')
```

```python
sns.heatmap(x, cbar=False, linewidths=0.5, annot=True, fmt='d', cmap='crest')
```
