# Violin Plots

### violin plots are a method of plotting numeric data and can be considerd a combination of the box plot with a kernel density plot.

### The unquestionable advantage of the violin plot over the box plot is that aside from showing the abovementioned ststistics it also shows the entire distribution of the data.

### Violin plots are often used to compare the distribution of a given variable across some categories.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.violinplot(tips['total_bill'])
```

```python
sns.catplot(x='day', y='total_bill', kind='violin', data=tips)
```

```python
sns.catplot(x='day', y='total_bill', kind='violin',hue='sex',split=True ,data=tips)
```

# Barplot and Countplot

```python
sns.catplot(x='smoker', y='total_bill', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex', kind='bar', data=tips)
```

```python
sns.catplot(x='smoker', y='total_bill', hue='sex',estimator=np.var, kind='bar', data=tips)
```

```python
sns.catplot(x='sex',hue='smoker', kind='count', data=tips)
```
