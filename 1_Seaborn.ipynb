## Scatter Plots
### The scatter plot is a mainstay of statistical visualization. It depends the joint distribution of two variables using a cloud of points, where each point represents an observation in the dataset. This depiction allows the eye to infer a substantial amount of information about whether there is any meaningful relationship between them.

#### There are 2 ways to draw a scatterplot in seaborn

#### 1. relplot [Figure Level Function]
#### 2. scatterplot [Axes Level Function]

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

## Plot a Scatter Plot between total_bill and tip

```python
# Plot here
```

```python
sns.relplot(x='total_bill', y='tip',kind='scatter', data=data)
```

## The hue parameter 

```python
# Plot here
sns.relplot(x='total_bill', y='tip', hue='smoker', data=data)
```

## The style parameter

```python
# Plot here
sns.relplot(x='total_bill', y='tip', hue='smoker', style='sex',data=data)
```

## The size parameter 

```python
# Plot Here
sns.relplot(x='total_bill', y='tip', hue='smoker',style='sex',size='size',sizes=(15,200),data=data)
```

## Using the scatterplot function

```python
# Plot here
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=data)
```
