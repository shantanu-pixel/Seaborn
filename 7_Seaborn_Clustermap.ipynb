```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('fivethirtyeight')
```

```python
flights=sns.load_dataset('flights')
```

```python
x=flights.pivot_table(index='year', columns='month'
 ,values='passengers', aggfunc='sum')
```

## Clustermap 

```python
sns.clustermap(x, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True,row_cluster=False, metric='correlation')
```

```python
sns.clustermap(x, z_score=0, annot=True,col_cluster=False, metric='correlation')
```
